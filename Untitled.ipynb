```python
tail_positions = set()

tail_position = head_position = [0,0]

def move_head(head_position, direction):
    directions = {
    "U": [0, 1],
    "D": [0, -1],
    "L": [-1, 0],
    "R": [1, 0],
    }
    return [dim1 + dim2 for dim1, dim2 in zip(head_position, directions[direction])]

for line in open("Day9_input.txt", "r"):
    for i in range(int(line.strip("\n")[-1])):
        print(head_position, tail_position)
        head_position = move_head(head_position, line.strip("\n")[0])
        
        if (abs(tail_position[0] - head_position[0]) > 1 and abs(tail_position[1] - head_position[1]) >= 1) \
        or (abs(tail_position[1] - head_position[1]) > 1 and abs(tail_position[0] - head_position[0]) >= 1):
            tail_position[0] = tail_position[0] + (head_position[0] - tail_position[0]) / abs((head_position[0] - tail_position[0]))
            tail_position[1] = tail_position[1] + (head_position[1] - tail_position[1]) / abs((head_position[1] - tail_position[1]))
        
        elif abs(tail_position[0] - head_position[0]) > 1:
            tail_position[0] = tail_position[0] + (head_position[0] - tail_position[0]) / abs((head_position[0] - tail_position[0]))
            
        elif abs(tail_position[1] - head_position[1]) > 1:
            tail_position[1] = tail_position[1] + (head_position[1] - tail_position[1]) / abs((head_position[1] - tail_position[1]))
            
        tail_positions.add(tuple(tail_position))
```
