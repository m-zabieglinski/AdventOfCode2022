```python
H_movement = []

directions = {
    "U": [0, 1],
    "D": [0, -1],
    "L": [-1, 0],
    "R": [1, 0],
}

for line in open("Day9_input.txt", "r"):
    for i in range(int(line.strip("\n").split()[-1])):
        H_movement.append(directions[line.strip("\n").split()[0]])

H_movement
```

```python
H_positions = [[0, 0]]
T_positions = [[0, 0]]

def vector_sum(vec1, vec2):
    return [dim1 + dim2 for dim1, dim2 in zip(vec1, vec2)]

def vector_difference(vec1, vec2):
    return [dim1 - dim2 for dim1, dim2 in zip(vec1, vec2)]

def chebyshev_distance(vec1, vec2):
    return max([abs(dim) for dim in vector_difference(vec1, vec2)])

for move in H_movement:
    
    H_positions.append(vector_sum(H_positions[-1], move))
    
    if chebyshev_distance(H_positions[-1], T_positions[-1]) > 1:
        T_positions.append(H_positions[-2])
        
    else:
        T_positions.append(T_positions[-1])
        
T_positions = [(int(pos[0]), int(pos[1])) for pos in T_positions]
H_positions = [(int(pos[0]), int(pos[1])) for pos in H_positions]      
len(set(T_positions))
```
