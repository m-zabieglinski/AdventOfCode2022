```python
import re
sensor_beacon = {}
for line in open("Day15_input.txt", "r"):
    sensor_part, beacon_part = line.split(":")
    sensor = tuple([int("".join(re.findall("-|\d", coord))) for coord in sensor_part.split(",")])
    beacon = tuple([int("".join(re.findall("-|\d", coord))) for coord in beacon_part.split(",")])
    sensor_beacon[sensor] = beacon
sensor_beacon
```

```python
def get_distance(p1, p2):
    return (abs(p1[0] - p2[0]) + abs(p1[1] - p2[1]))

def get_edge_points_covered(sensor, y):
    d = get_distance(sensor, sensor_beacon[sensor])
    y_b, x_b = sensor[1], sensor[0]
    x1 = -d + abs(y_b - y) + x_b
    x2 = d - abs(y_b - y) + x_b
    return x1, x2

def get_all_points_covered(sensor, y):
    x1, x2 = get_edge_points_covered(sensor, y)
    points = []
    for point in range(x1, x2+1):
        points.append(point)
    return points
```

```python
points_for_all_sensors = []

for sensor in sensor_beacon.keys():
    points_for_all_sensors += get_all_points_covered(sensor, 2000000)
    
len(set(points_for_all_sensors)) - 1
```
